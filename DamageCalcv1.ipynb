```python
import json
```

```python
def load_character(filepath):
    with open(filepath, "r") as f:
        data = json.load(f)
    return data
def get_normal_attack_multiplier(character_data, hit_number):
    hits = character_data["talents"]["normal_attack"]["hits"]
    for hit in hits:
        if hit["hit_number"] == hit_number:
            return hit["multiplier"]
    return None
def calculate_base_damage(multiplier,scaling_stat_value):
    return (multiplier / 100) * scaling_stat_value
def apply_defense_multiplier(damage, enemy_def, character_level):
    def_multiplier = (character_level + 100) / (enemy_def + character_level + 100)
    return damage * def_multiplier
def calculate_hit_damage(character_data, hit_number, enemy_def, character_level):
    multiplier = get_normal_attack_multiplier(character_data, hit_number)
    scaling_stat_name = character_data["talents"]["normal_attack"]["scaling_stat"]
    scaling_stat_value = character_data["base_stats"][scaling_stat_name]
    base = calculate_base_damage(multiplier, scaling_stat_value)
    final = apply_defense_multiplier(base, enemy_def, character_level)
    return final
```

```python
navia_data=load_character("navia.json")
hit_number = navia_data["talents"]["normal_attack"]["hits"]
total = 0
for hit in hit_number:
        result = calculate_hit_damage(navia_data, hit["hit_number"], enemy_def=100, character_level=90)
        total = total + result
print(total)
```
